```python
print(df.columns)
```

```python
# Create new features

# Average installment amount (assuming installment represents periodic payments)
df['avg_installment_amount'] = df['installment']

# Total payment ratio: total payments made / funded amount
df['total_payment_ratio'] = df['total_pymnt'] / df['funded_amnt']

# Interest received ratio: total interest received / funded amount
df['interest_received_ratio'] = df['total_rec_int'] / df['funded_amnt']

# Recoveries ratio: recoveries / funded amount
df['recoveries_ratio'] = df['recoveries'] / df['funded_amnt']
```

```python
# Display the first few rows to verify the new features
print(df.head())
```

```python
# Generate interaction features between existing variables, like income-to-loan 
# ratio.
```

```python
# Income-to-Loan Ratio
df['income_to_loan_ratio'] = df['annual_inc'] / df['loan_amnt']

# Debt-to-Income Ratio
df['debt_to_income_ratio'] = df['dti'] / df['annual_inc']

# Installment-to-Income Ratio
df['installment_to_income_ratio'] = df['installment'] / df['annual_inc']

# Loan Amount-to-Funded Amount Ratio
df['loan_to_funded_ratio'] = df['loan_amnt'] / df['funded_amnt']

# Revolving Balance-to-Income Ratio
df['revol_balance_to_income_ratio'] = df['revol_bal'] / df['annual_inc']
```

```python
print(df.head())
```

```python
# Use domain knowledge to create features that might influence loan defaults, 
# such as employment stability and credit score.
```

```python
# Employment Length: Convert employment length to a numerical value (assuming it's in a string format)
df['emp_length_num'] = df['emp_length'].str.extract('(\d+)').astype(float)
df['emp_length_num'].fillna(0, inplace=True)  # Fill NaN values with 0 (e.g., 'n/a' or '< 1 year')

# Is Long-Term Employed: Binary feature if employed for more than 10 years
df['is_long_term_employed'] = df['emp_length_num'] > 10

# Revolving Balance Utilization
df['revol_bal_util'] = df['revol_bal'] / df['total_rev_hi_lim']

# Loan to Income Ratio
df['loan_to_income_ratio'] = df['loan_amnt'] / df['annual_inc']

# Number of Inquiries in the last 6 months
df['num_inquiries_last_6m'] = df['inq_last_6mths']

# Number of Inquiries in the last 12 months
df['num_inquiries_last_12m'] = df['inq_last_12m']

# Number of Open Accounts
df['num_open_accounts'] = df['open_acc']
```
