**Question 1**

```python
df['patient_id'] = df.index // 5
```

```python
df['readmission'] = df['readmitted'].apply(lambda x: 1 if x == 'yes' else 0)
```

```python
df['prior_readmissions'] = df.groupby('patient_id')['readmission'].cumsum() - df['readmission']
```

```python
df.head()
```

**Question 2**

```python
# Convert the age to numberical format
age_map = {'[0-10)':5,'[10-20)':15,'[20-30)':25,'[30-40)':35,'[40-50)':45,'[50-60)':55,'[60-70)':65,'[70-80)':75,'[80-90)':85,'[90-100)':95,}
df['age_num'] = df['age'].map(age_map)
```

```python
df['age_num'].head()
```

```python
df['age_prior_readmissions'] = df['age_num'] * df['prior_readmissions']
```

```python
# converting the age into groups
df['age_groups'] = pd.cut(df['age_num'] ,bins=[0,18,35,50,65,80,100], labels=['0-18', '19-35', '36-50', '51-65', '66-80', '81+'])
```
